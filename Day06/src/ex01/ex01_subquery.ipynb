## Разрешенные функции : импорт pandas как pd, импорт sqlite3

```python
import pandas as pd 
import sqlite3
```

## создайте подключение к базе данных с помощью библиотеки sqlite3
## получите схему средства проверки checker

```python
conn=sqlite3.connect('../data/checking-logs.sqlite')
pd.read_sql('pragma table_info(checker)',conn,index_col='cid')
```

## для проверки таблицы используйте только первые 10 строк

```python
query="""
select *
from checker
limit 10
"""

pd.read_sql(query,conn,index_col='index')
```

## Подсчитайте, сколько строк удовлетворяют следующим условиям, используя только один запрос с любым количеством подзапросов:
### - посчитайте количество строк в таблице просмотров страниц, но только с пользователями из таблицы проверки:
### - статус = «готов», мы не хотим анализировать журналы, которые находятся в режиме проверки статуса
### - numTrials = 1, мы хотим проанализировать только первые коммиты, потому что только они могут подсказать нам, когда студент начал работать над лабораторной работой
### - Названия лабораторий должны быть из списка: ’laba04’, ’laba04s’, ’laba05’, ’laba06’, ’laba06s’, ’project1’. Только они были активны во время эксперимента
### храните в фрейме данных шашки со столбцом cnt

```python
query = """
        select count(*) as cnt
        from pageviews
        where uid in (
        select uid from checker
        where status = 'ready' and numTrials = 1
        and labname in ('laba04', 'laba04s', 'laba05', 'laba06', 'laba06s', 'project1')
        )
        """
pd.read_sql(query,conn)
```
