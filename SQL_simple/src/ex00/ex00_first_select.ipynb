### Разрешенные функции : импорт pandas как pd, импорт sqlite3

```python
import pandas as pd
import sqlite3
```

### создайте подключение к базе данных с помощью библиотеки sqlite3

```python
conn=sqlite3.connect('../data/checking-logs.sqlite')
```

### получите схему таблицы просмотров страниц с помощью pd.io.sql.read_sql и запроса «PRAGMA table_info(pageviews);»

```python
schema=pd.read_sql('PRAGMA table_info(pageviews);',conn)
schema
```

### для просмотра таблицы достаточно первых 10 строк

```python
pd.read_sql('select* from pageviews limit 10',conn,index_col='index')
```

### получите подтаблицу, используя только один запрос, где:
### * используются только uid и datetime

```python
pd.read_sql('select uid, datetime from pageviews limit 10',conn)
```

### используются только данные пользователя (user_ *), а не данные администратора

```python
query="""
select uid, datetime
from pageviews
where uid like 'user_%'
"""

pd.read_sql(query,conn)
```

### он отсортирован по uid в порядке возрастания

```python
query="""
select uid, datetime
from pageviews
where uid like 'user_%'
order by uid asc
"""

pd.read_sql(query,conn)
```

### столбец индекса - datetime
### datetime преобразуется в DatetimeIndex
### имя фрейма данных - просмотры страниц

```python
query="""
select uid, datetime
from pageviews
where uid like 'user_%'
order by uid asc
"""

pageviews = pd.read_sql(query, conn, index_col='datetime')
pageviews.index = pd.to_datetime(pageviews.index)
pageviews
```
