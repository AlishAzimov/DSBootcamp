## обогатите фрейм данных, используя образец из этого фрейма данных

### создайте выборку из 200 новых наблюдений с помощью random_state = 21
### в выборке не должно быть новых комбинаций car number, make и model, поэтому весь набор данных будет согласован по этим параметрам
### нет ограничений на refund и fines, вы можете взять любое значение из этих столбцов наугад и использовать его для любого номера автомобиля
### объедините образец с исходным фреймом данных в новый фрейм данных concat_rows

```python
sample_df = df.sample(n=200, random_state=21)
sample_df
```

```python
con_cat_rows = pd.concat([df, sample_df], ignore_index=True)

con_cat_rows
```

## обогатите фрейм данных concat_rows новым столбцом с сгенерированными данными

### создайте серию с именем Year, используя случайные целые числа от 1980 до 2019
### используйте np.random.seed(21) перед созданием лет
### объедините серию с фреймом данных и назовите ее fines

```python
np.random.seed(21)

con_cat_rows['Year'] = [np.random.randint(1980, 2019) for i in range(len(con_cat_rows))]

con_cat_rows
```
