## Работа с пропущенными значениями

### проверьте, сколько значений отсутствует в каждом столбце
### удалите все столбцы, в которых более 500 пропущенных значений, используя аргумент thresh, проверьте, сколько пропущенных значений в каждом столбце
### замените все пропущенные значения в столбце Refund предыдущим значением в этом столбце для этой ячейки, используйте метод аргументов, проверьте, сколько значений пропущено в каждом столбце
### замените все пропущенные значения в столбце Fines на среднее значение этого столбца (при вычислении среднего значения исключите значения NA/null), проверьте, сколько значений пропущено в каждом столбце

```python
df.isnull().sum()
```

```python
df.dropna(axis=1, thresh=500, inplace=True)
df.isnull().sum()
```

```python
df['Refund'] = df['Refund'].ffill()
df.isnull().sum()
```

```python
mean = df['Fines'].mean()
df['Fines'] = df['Fines'].fillna(mean)
df.isnull().sum()
```

## Разделите и проанализируйте марку и модель

### используйте метод apply как для разделения, так и для извлечения значений в новые столбцы Make и Model

### отбросьте колонку Make_n_model

### сохраните фрейм данных в JSON-файле auto.json в приведенном ниже формате:

```python
df["Make"] = df["Make_n_model"].apply(lambda x: str(x).split(" ")[0])
df["Model"] = df["Make_n_model"].apply(lambda x: " ".join(str(x).split(" ")[1:]))
df
```

```python
df.drop(columns=["Make_n_model"], inplace=True)
df
```

```python
df.to_json("../../datasets/auto.json", orient="records", indent=2)
```
