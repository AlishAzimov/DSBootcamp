## Cоздайте фрейм данных views с двумя столбцами: datetime и user, прочитав feed-views.log

```python
views=pd.read_csv('../../datasets/feed-views.log', names=['datetime', 'user'], sep='\t')
views.info(memory_usage='deep')
```

```python
views
```

### Преобразуйте datetime в datetime64[ns] Dtype

```python
views['datetime']=views['datetime'].astype('datetime64[ns]')
views.info(memory_usage='deep')
```

### Извлеките год, месяц, день, час, минуту и секунду из значений этого столбца в новые столбцы

```python
views['year']=views['datetime'].dt.year
views['month']=views['datetime'].dt.month
views['day']=views['datetime'].dt.day
views['hour']=views['datetime'].dt.hour
views['minute']=views['datetime'].dt.minute
views['second']=views['datetime'].dt.second
views['microsecond']=views['datetime'].dt.microsecond
display(views.info(memory_usage='deep'))
views.head()
```

## Cоздайте новый столбец daytime

### Вам нужно присвоить конкретное значение времени суток, если час находится в определённом интервале, например, полдень, если час больше 11 и меньше или равен 17
### 0 – 3,59 night, 4 – 6,59 early morning, 7 – 10,59 morning, 11 – 16,59 afternoon, 17 – 19,59 early evening, 20 – 23,59 evening
### используйте метод cut для решения этой подзадачи
### назначьте столбец user в качестве индекса

```python
views['daytime'] = pd.cut(views['hour'],
                           bins=[0, 4, 7, 11, 17, 20, 24], 
                           labels=['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening'], right=False)
views
```

### Отсортируйте значения в вашем фрейме данных по часам, минутам и секундам в порядке возрастания (одновременно, а не по одному)

## Рассчитайте минимум и максимум для часов и режима для категорий дневного времени

### рассчитайте максимальное количество часов для строк, где время суток равно ночи
### рассчитайте минимальное количество часов для строк, где время суток - утро
### Кроме того, выясните, кто посещал страницу в эти часы (приведите один пример)
### рассчитайте mode для часа и дневного времени

```python
night_df = views[views['daytime'] == 'night']
morning_df = views[views['daytime'] == 'morning']

max_night = night_df['hour'].max()
min_morning = morning_df['hour'].min()

print(f'Max hour night: {max_night}')
print(f'Min hour morning: {min_morning}')


users_night = views[views['hour'] == max_night]
users_morning = views[views['hour'] == min_morning]

print(users_night)
print(users_morning)

print(f'Moda for hour: {views['hour'].mode()}')
print(f'Moda for daytime: {views['daytime'].mode()}')
```

## Используйте метод describe() для получения базовой статистики по столбцам

### чтобы узнать, какой интервал посещения страницы является самым популярным, вычислите межквартильный размах для часа, извлекая значения из результата метода describe() и сохраняя их в переменной iqr


```python
desc = views.describe()
display(desc)
q1=desc['hour']['25%']
q3=desc['hour']['75%']
print(f'IQR = {q3-q1}')
```
