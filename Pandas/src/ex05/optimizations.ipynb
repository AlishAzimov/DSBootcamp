## итерации: во всех следующих подзадачах вам нужно вычислить **fines/refund*year** для каждой строки и создать новый столбец с вычисленными данными, а также измерить время с помощью магической команды %%timeit в ячейке

### цикл: напишите функцию, которая перебирает данные в фрейме данных с помощью for i in range(0, len(df)), iloc и append() и сохраняет их в списке. Присвойте результат работы функции новому столбцу в фрейме данных
### сделайте это с помощью iterrows()
### сделайте это с помощью apply() и лямбда-функции
### сделайте это, используя Series объекты из фрейма данных
### сделайте это, как в предыдущей подзадаче, но с помощью метода .values

```python
def loop(frame):
    data = []
    for i in range(0, len(frame)):
        row = df.iloc[i] 
        value = row['Fines'] / row['Refund'] * row['Year']
        data.append(value)
    return data
```

```python
%%timeit
df['Result'] = loop(df)
```

```python
def iter_row(frame):
    data = []
    for _, row in frame.iterrows():
        value = row['Fines'] / row['Refund'] * row['Year']
        data.append(value)
    return data
```

```python
%%timeit
df['Result'] = iter_row(df)
```

```python
%%timeit
df['Result'] = df.apply(lambda row: row['Fines'] / row['Refund'] * row['Year'], axis=1)
```

```python
%%timeit
df['Result'] = df['Fines'] / df['Refund'] * df['Year']
```

```python
%%timeit
df['Result'] = df['Fines'].values / df['Refund'].values * df['Year'].values
```

### запустите df.info(memory_usage=’deep’), обратите внимание на Dtype и использование памяти
### преобразуйте copy() ваш исходный фрейм данных в другой фрейм данных optimized
### уменьшенный от float64 до float32 для всех столбцов
### уменьшите значение с int64 до минимально возможного числового dtype
### запустите info(memory_usage='deep') для вашего нового фрейма данных, обратите внимание на тип данных и использование памяти

```python
df.info(memory_usage='deep')
```

```python
df_optimized = df.copy()

df_optimized['Refund'] = df_optimized['Refund'].astype('int8')
df_optimized['Fines'] = df_optimized['Fines'].astype('float32')
df_optimized['Year'] = df_optimized['Year'].astype('int16')
df_optimized['Result'] = df_optimized['Result'].astype('float32')
```

```python
df_optimized.info(memory_usage='deep')
```
