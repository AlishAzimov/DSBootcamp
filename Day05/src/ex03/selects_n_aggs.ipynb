## создайте агрегированные данные с указанием марки и модели

### отобразить медиану, fines сгруппированную по make
### отобразить медиану, fines сгруппированную по make и model
### отобразите количество fines сгруппированных по make и model для того, чтобы понять, можно ли доверять медианным значениям
### отобразите минимальное и максимальное значения fines в группах make и model для лучшего понимания дисперсии
### отобразите стандартное отклонение fines сгруппированных по make и model для лучшего понимания дисперсии

```python
df['Fines'].groupby(df['Make']).median()
```

```python
df['Fines'].groupby(by=[df['Make'],df['Model']]).median()
```

```python
df['Fines'].groupby(by=[df['Make'],df['Model']]).count()
```

```python
df['Fines'].groupby(by=[df['Make'],df['Model']]).agg(['min', 'max'])
```

```python
df['Fines'].groupby(by=[df['Make'],df['Model']]).std()
```

```python
df['Fines'].groupby(by=[df['Make'],df['Model']]).agg(['mean','median','count','min', 'max','std'])
```
